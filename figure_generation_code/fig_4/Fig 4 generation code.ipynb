```python
mlp_decorr_1_bacc_res = open_pickle('Data/mlp_decorr_1_bacc_res.pkl')
mlp_early_stopping_decorr_2_bacc_res = open_pickle('Data/mlp_early_stopping_decorr_2_bacc_res.pkl')

svc_c10k_bacc_res_decorr_1 = open_pickle('Data/svc_c10k_bacc_res_decorr_1.pkl')
svm_decorr_2_bacc_res = open_pickle('Data/svm_decorr_2_bacc_res.pkl')
```

```python
fig_height = 1050
fig_width = 4474
enlarge = 350

fig, axs = plt.subplots(1,2, figsize=(int(fig_width/enlarge), int(fig_height/enlarge)), dpi=96)


ax = axs[0]
bar_plot([mlp_bacc_res[0], mlp_decorr_1_bacc_res[0], mlp_early_stopping_decorr_2_bacc_res[0]], ["Original", "Mean = 0", "Mean = 0, Cov = I"], ax=ax, ylabel="Balanced accuracy", 
         ylim=[.40, 1], bottom=.495, show=False, paired="condition", show_sample_size=False, sig_arr=[(7,8)], sig_th=.05, show_legend=False, show_stat=False, show_data=True, 
         show_lines=True)
title_plot("NN", (.5, 1.2), ax=ax, show_data=True)
all_data = sum([list(x) for x in zip(*[mlp_bacc_res[0], mlp_decorr_1_bacc_res[0], mlp_early_stopping_decorr_2_bacc_res[0]])], [])
draw_stats(all_data, ax, base=.5)


ax = axs[1]
bar_plot([svc_c10k_bacc_res[0], svc_c10k_bacc_res_decorr_1[0], svm_decorr_2_bacc_res[0]], ["Original", "Mean = 0", "Mean = 0, Cov = I"], ax=ax, ylabel="Balanced accuracy", ylim=[.40, 1], 
         bottom=.495, show=False, paired="condition", show_sample_size=False, sig_arr=[(7,8)], sig_th=.1, show_legend=True, show_stat=False, show_data=True, 
         show_lines=True, sig_symbol="~")
         
title_plot("gSVM", (.5, 1.2), ax=ax, show_data=True)
all_data = sum([list(x) for x in zip(*[svc_c10k_bacc_res[0], svc_c10k_bacc_res_decorr_1[0], svm_decorr_2_bacc_res[0]])], [])
ax.set_yticks([.5, .6, .7, .8, .9, 1.])
for i, g in enumerate(all_data):
    margin = .04
    h = np.mean(g) + np.std(g)/np.sqrt(len(g)) + margin
    ix = i + i // 3 * .5
    x,y = d2ax(ax, np.array([(ix,h)]))[0]
    tres = ttest_1samp(g, 0.5)
    _,p = tres
    p *= 1
    symbol = "n.s."
    if p < .05:
        symbol = "*"
    if p < .01:
        symbol = "**"
    if p < .001:
        symbol = "***"
    ax.text(x, y, symbol, fontsize = 12, horizontalalignment="center", verticalalignment="center", clip_on=False, transform=ax.transAxes)
#plt.savefig('fig 4f.svg', dpi=fig.dpi)
```
